```python
import triplespec_spectroimages as tspec

####################################
infolder  = 'D:/SOAR/TSpec/UT20220730/'
outfolder = infolder + 'IRAS_spec2d_python/'

# results from XSpexTool Pipeline
flat_solution  = infolder + 'CALnew/f.fits' # IDL output from XSpexTool > Cals > Create Flat
arc_solution   = infolder + 'CALnew/a.fits' # IDL output from XSpexTool > Cals > Wavelength Calibration
flux_solution  = infolder + 'PROC_wd1/IRAS16475_tcor_223_tellspec.fits' # IDL output from XTellCorr

# Raw images
science_frame = infolder + 'RAW/SPEC_IRAS16475_0239.fits' # Raw science frame
sky_frame     = infolder + 'RAW/SPEC_IRAS16475_0238.fits' # Raw sky frame, set to 'None' if no sky was observed

sky_ratio = 1.005 # Multiplicative factor for improving the sky subtraction by scaling the sky frame (default = 1.0)
plot_ci   = 0.95  # confidence interval for plotting the resulting spectroimage (useful for removing outliers)

# Create the spectro images
tspec.tspec_extract2d(science_frame, flat_solution, arc_solution, outfolder,
                      sky_frame=sky_frame, sky_ratio=sky_ratio,
                      flux_solution=flux_solution, vhelio_correction=True,
                      reverse_slit=False,
                      plot_aspectratio=4, plot_ci=plot_ci, plot_dpi=300)
```
